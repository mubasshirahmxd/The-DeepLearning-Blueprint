```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing.image import img_to_array, load_img

# Create an image generator with data augmentation
datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.4,
    height_shift_range=0.4,
    shear_range=0.4,
    zoom_range=0.4,
    horizontal_flip=True,
    rescale = None,
    fill_mode='nearest'
)
```

```python

# Load the image
img = load_img(r"C:\Users\MUBASSHIR\OneDrive\Desktop\FSDS-Class-Files\data_augmentation\2.jpg")
```

```python
img
```

```python
# convert the image to a numpy array
x = img_to_array(img)
x = x.reshape((1,) + x.shape)  # reshape to (1, height, width, channels)

# Generate and save augmented images
i = 0
for batch in datagen.flow(x, batch_size=1,
                          save_to_dir=r"C:\Users\MUBASSHIR\OneDrive\Desktop\FSDS-Class-Files\data_augmentation\exampleFolder", save_prefix='dog', save_format='jpg'):
    i += 1
    if i > 20:  # Generate 20 augmented images
        break   
```

## üéØ Objective
This notebook demonstrates **basic image handling operations in OpenCV** ‚Äî reading, displaying, and saving images.  
We'll also explore color spaces and visualization differences between **cv2** and **matplotlib**.

---



## 1Ô∏è‚É£ Importing Libraries
Let's import the required libraries and check OpenCV version.


````python

import cv2
import matplotlib.pyplot as plt

print("# ‚û§ OpenCV version:", cv2.__version__)
````


---

## 2Ô∏è‚É£ Reading an Image
Use `cv2.imread()` to read an image from a file path.

**Syntax:**
```python
cv2.imread(filename, flag)
```
- `filename`: Path to the image.
- `flag`: 
  - `1` (default) ‚Üí Read in color (BGR)
  - `0` ‚Üí Grayscale
  - `-1` ‚Üí Unchanged (including alpha transparency)


````python

# Read image from folder
img_path = 'sample_images/elephant.jpeg'

# Read in color (default)
image = cv2.imread(img_path, 1)

# Check if image loaded correctly
if image is None:
    print("‚ùå Error: Image not found or path incorrect!")
else:
    print("‚úÖ Image loaded successfully!")
    print("# ‚û§ Image shape:", image.shape)  # (Height, Width, Channels)
````


---

## 3Ô∏è‚É£ Displaying Image (Local Environment)
In a local Python environment (VS Code, PyCharm, etc.), use `cv2.imshow()`:

```python
cv2.imshow('Elephant', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

‚ö†Ô∏è Note: `cv2.imshow()` **won‚Äôt work in Jupyter/Colab** ‚Äî it needs a GUI window.

Hence, we‚Äôll use **Matplotlib** for display instead.


````python
cv2.imshow('Elephant', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
````


---

## 4Ô∏è‚É£ Displaying Image using Matplotlib
Matplotlib expects **RGB** images, but OpenCV loads them as **BGR**,  
so colors appear inverted if we don‚Äôt convert them.

Let‚Äôs fix that using `cv2.cvtColor()`.


````python

# Convert BGR to RGB for correct color display
img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display using matplotlib
plt.figure(figsize=(6,6))
plt.imshow(img_rgb)
plt.title('Elephant Image (RGB Corrected)')
plt.axis('off')
plt.show()
````


---

## 5Ô∏è‚É£ Writing (Saving) Images
Use `cv2.imwrite()` to save images to disk.

**Syntax:**
```python
cv2.imwrite('new_image_name.jpg', image)
```


````python

# Save the image in a new file
output_path = 'sample_images/elephant_copy.jpg'
cv2.imwrite(output_path, image)

print("‚úÖ Image saved successfully as:", output_path)
````


---

## 6Ô∏è‚É£ Convert to Grayscale and Display
Let‚Äôs also convert our image to grayscale and display it.


````python

# Convert to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(6,6))
plt.imshow(gray_image, cmap='gray')
plt.title('Elephant Image (Grayscale)')
plt.axis('off')
plt.show()
````


---

## üß† Summary of Learned Functions

| Function | Description |
|-----------|--------------|
| `cv2.imread(path, flag)` | Reads image from disk |
| `cv2.imshow(name, image)` | Displays image in new window (local) |
| `cv2.waitKey(0)` | Waits for key press to close window |
| `cv2.destroyAllWindows()` | Closes all OpenCV windows |
| `cv2.cvtColor(image, code)` | Converts color space (BGR‚ÜîRGB‚ÜîGRAY) |
| `cv2.imwrite(path, image)` | Saves image to disk |

‚úÖ You now know how to **read, display, and save** images in OpenCV!
